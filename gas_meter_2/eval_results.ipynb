```python
import os, json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
params = []
for fn in os.listdir("results"):
    if fn.endswith("json"):
        with open(os.path.join("results", fn), "r") as file:
            params.append(json.load(file))
            
df = pd.DataFrame.from_records(params)
df.head()
```

```python
df["pass"] = [{True: 1, False:0}[ocr] for ocr in df.ocr_plausible]
df.head()
```

```python
df.describe()
```

```python
df.plot(kind="scatter", y="ascpect_ratio", x="rect_factor", c="pass", cmap=plt.cm.rainbow);
```
