```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Read image

```python
img_raw = cv.imread('gas_sample.jpg', cv.IMREAD_COLOR)
img_raw = cv.cvtColor(img_raw, cv.COLOR_BGR2RGB)  # opencv reads colors in BGR
plt.imshow(img_raw);
```

## Scale image

```python
def rescale_image(image, scale=0.75):
    width = int(image.shape[1] * scale)
    height = int(image.shape[0] * scale)
    return cv.resize(image, (width, height), interpolation=cv.INTER_AREA)
```

```python
img_mini = rescale_image(img_raw, 0.2)
plt.imshow(img_mini);
```

## Convert color to grayscale

```python
img_gray = cv.cvtColor(img_mini, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray, plt.cm.gray);
```

# Ignore red color

```python
img_bgr = cv.imread('gas_sample.jpg', cv.IMREAD_COLOR)
def bgr_2_gray(img):
    """ Converts a BGR image into a grayscale image based on blue and green, ignoring red 
    """
    return np.array(img[:,:,:2].mean(axis=2), dtype=np.uint8)

img_bg_gray = bgr_2_gray(img_bgr)
```
