# Invert an Image

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_bgr = cv.imread('gas_sample.jpg', cv.IMREAD_COLOR)
img_raw = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY)
plt.imshow(img_raw, plt.cm.gray);
```

```python
img_inv = cv.bitwise_not(img_raw)
plt.imshow(img_inv, plt.cm.gray);
```
