## Thresholding
https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_bgr = cv.imread('gas_sample.jpg', cv.IMREAD_COLOR)
img_raw = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY)
plt.imshow(img_raw, plt.cm.gray);
```

1. Manual Threshold

```python
_, img_bin = cv.threshold(img_raw, thresh=100, maxval=255, type=cv.THRESH_BINARY)
plt.imshow(img_bin, plt.cm.gray);
```

2. OTSU thresholding with automatic threshold selection

```python
_, img_otsu = cv.threshold(img_raw, 0, 255, cv.THRESH_OTSU)
plt.imshow(img_otsu, plt.cm.gray);
```

3. Adaptive thresholds

```python
img_adp = cv.adaptiveThreshold(img_raw, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 10)
plt.imshow(img_adp, plt.cm.gray);
```
