The function cv.floodFill fills a connected component starting from the seed point with the specified color.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('coin.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img, plt.cm.gray);
```

```python
_, im_th = cv.threshold(img, 220, 255, cv.THRESH_BINARY_INV)
plt.imshow(im_th, plt.cm.gray);
```

```python
im_floodfill = im_th.copy()

# Mask used to flood filling.
# Notice the size needs to be 2 pixels than the image.
h, w = im_th.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)
cv.floodFill(im_floodfill, mask, (0,0), 255)
plt.imshow(im_floodfill, plt.cm.gray);
```

```python
im_floodfill_inv = cv.bitwise_not(im_floodfill)
plt.imshow(im_floodfill_inv, plt.cm.gray);
```

```python
im_out = im_th | im_floodfill_inv
plt.imshow(im_out, plt.cm.gray);
```
