# Perspective transfromation with opencv
Code take from https://docs.opencv.org/4.x/da/d6e/tutorial_py_geometric_transformations.html

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
name1 = 'grid_distorted.png'
img1 = cv.imread(name1, cv.IMREAD_COLOR)
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)  # opencv reads colors in BGR
img1.shape
```

```python

pts1 = np.float32([[243, 105], [1436, 64], [88, 866], [1215, 959]])
w2, h2 = 1200, 1000
pts2 = np.float32([[0, 0], [w2, 0], [0, h2], [w2, h2]])

matrix = cv.getPerspectiveTransform(pts1, pts2)

img2 = cv.warpPerspective(img1, matrix,(w2, h2))
```

```python
plt.subplot(121),plt.imshow(img1),plt.title(name1)
plt.subplot(122),plt.imshow(img2),plt.title('grid transformed')
plt.show()
```

**Even rotation is possible by swapping the points...**

```python
pts3 = np.float32([[w2, h2], [0, h2], [w2, 0], [0, 0]])
matrix = cv.getPerspectiveTransform(pts1, pts3)
img3 = cv.warpPerspective(img1, matrix,(w2, h2))
plt.imshow(img3), plt.title("rotated");
```
