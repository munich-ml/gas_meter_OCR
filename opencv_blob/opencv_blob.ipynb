# Blob detection
https://learnopencv.com/blob-detection-using-opencv-python-c/

```python
import pytesseract
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
img = cv.imread('blobs.png', cv.IMREAD_GRAYSCALE)
plt.imshow(img, plt.cm.gray);
```

```python
detector = cv.SimpleBlobDetector_create()
keypoints = detector.detect(img)
img_kp = cv.drawKeypoints(img, keypoints, np.array([]), (0, 255, 0), 
                          cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_kp, plt.cm.gray);
for i, kp in enumerate(keypoints):
    print(f"{i=}, {kp.pt=}, {kp.size=}")
```

## Try with gas-meter image

```python
img = cv.imread('gas_sample.jpg', cv.IMREAD_GRAYSCALE)
detector = cv.SimpleBlobDetector_create()
keypoints = detector.detect(img)
print(f"{len(keypoints)=}")
iwb = cv.drawKeypoints(img, keypoints, np.array([]), (0, 255, 0), 
                          cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(iwb, plt.cm.gray);
```

```python
params = cv.SimpleBlobDetector_Params()
```

```python
dct = {kp.size: kp for kp in keypoints}
largest_keys = sorted(dct.keys())[::-1][:3]
largest = [dct[key] for key in largest_keys]
iwb = cv.drawKeypoints(img, largest, np.array([]), (0, 255, 0), 
                          cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure(figsize=(15, 10))
plt.imshow(iwb, plt.cm.gray);
```
